```python
QUERY = ['P1', 'P2', 'P3',
          'P4', 'P3', 'P1',
          'P5', 'P2', 'P3',
            'P6', 'P7', 'P5',
            'P2', 'P1']
```

```python
from collections import OrderedDict, deque


class LRUCache:
    def __init__(self, capacity: int):
        self.cache = OrderedDict()
        self.capacity = capacity
        self.hits = 0

    def put(self, key, value):
        if key in self.cache:
            self.hits += 1
        self.cache[key] = value
        self.cache.move_to_end(key)
        if len(self.cache) > self.capacity:
            self.cache.popitem(last=False)  # 删除最不常用的（在字典开头的）

    def show_cache(self):
        print(f"Current cache (hits:{self.hits}):")
        for k, v in self.cache.items():
            print(f"{k}: {v}")

class FIFOCache:
    def __init__(self, capacity):
        self.cache = deque(maxlen=capacity)
        self.hits = 0
    def put(self, k, item):
        if item in self.cache:
            self.hits += 1
        else:
            self.cache.append(item)

    def show_cache(self):
        print(f"Current cache (hits:{self.hits}):")
        for v in list(self.cache):
            print(f"{v}")


class MRUCache:
    def __init__(self, capacity: int):
        self.cache = OrderedDict()
        self.capacity = capacity
        self.hits = 0

    def put(self, key, value):
        if key in self.cache:
            self.hits += 1
        self.cache[key] = value
        self.cache.move_to_end(key)
        if len(self.cache) > self.capacity:
            self.cache.popitem(last=True)  # 删除最常用的（在字典开头的）

    def show_cache(self):
        print(f"Current cache (hits:{self.hits}):")
        for k, v in self.cache.items():
            print(f"{k}: {v}")


CAPACITY = 4
# bufferpool = LRUCache(CAPACITY)
bufferpool = FIFOCache(CAPACITY)

for query in QUERY:
    v = query
    k = int(query.replace("P", ""))
    bufferpool.put(k, v)
    bufferpool.show_cache()



```
